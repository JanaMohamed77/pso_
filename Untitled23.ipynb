```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler
```

```python
data = np.array([1, 2, 3, 4, 5]).reshape(-1,1)
```

```python
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data)
```

```python
X = []
y = []

for i in range(len(data_scaled)-1):
    X.append(data_scaled[i])
    y.append(data_scaled[i+1])

X = np.array(X)
y = np.array(y)
```

```python
X = X.reshape((X.shape[0], 1, 1))
```

```python
model = Sequential()

model.add(LSTM(10, activation='tanh', input_shape=(1,1)))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(X, y, epochs=200, verbose=0)
```

```python
last_value = np.array([[data_scaled[-1]]])
last_value = last_value.reshape((1,1,1))

prediction = model.predict(last_value)
```

```python
predicted_value = scaler.inverse_transform(prediction)

print("Predicted next value:", predicted_value[0][0])
```
